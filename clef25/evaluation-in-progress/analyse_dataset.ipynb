```python
from ir_datasets_longeval import load
from upsetplot import from_contents
from upsetplot import UpSet
# !pip install upsetplot
```

```python
def prepare_data(dataset):
    dataset = load("longeval-sci/*")
    data = {}
    for d in dataset.get_datasets():
        queries = []
        for q in d.queries_iter():
            queries.append(q.query_id)
        data[d.get_snapshot()] = queries
    return data
```

```python
data = prepare_data("longeval-sci/*")
mysets = from_contents(data)
ax_dict = UpSet(mysets, subset_size="count", show_counts=True).plot()
```
